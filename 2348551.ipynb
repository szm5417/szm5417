```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
data = pd.read_csv('/content/2015.csv')
```

```
data.head()
```

```
data.shape
```

```
data.isnull().sum()
```

```
happiness_bins = [1.0,4.0,6.0,np.inf]
bins_label = ['Low Happiness','Medium Happiness','High happiness']
```

```
binned_happiness = pd.cut(data['Happiness Score'],bins= happiness_bins,labels=bins_label)
binned_happiness
```

```
data['binned_happiness']=binned_happiness
```

```
data.head()
```

```
data
```

```
encoded = pd.get_dummies(data,columns=['Region'])
```

```
encoded
```

```
from sklearn.preprocessing import MinMaxScaler
minmax = MinMaxScaler()
dataMIN = encoded['Economy (GDP per Capita)']
minmax.fit(dataMIN)
normalized = minmax.transform(dataMIN)
normalized
```

```
from scipy.stats import chi2_contingency
chi_test = pd.crosstab(data['Region'],data['Happiness Rank'])
chi,p,dof,expe = chi2_contingency(chi_test)
print('chi Value',chi)
print('p value',p)
print('dof - Degree of Freedoms',dof)
print('expected value',expe)
```

```
data_cov = encoded.cov()
```

```
data_cov
```

```
plt.figure(figsize=(15,15))
sns.heatmap(data_cov,annot=True,fmt=".2f", cmap="coolwarm",xticklabels=encoded.columns, yticklabels=encoded.columns)
plt.show()
```

```
plt.figure(figsize=(15,15))
sns.countplot(data,x=data['binned_happiness'],hue = data['Country'])
plt.show()
```
