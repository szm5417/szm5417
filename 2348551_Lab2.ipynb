```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/Womens Clothing E-Commerce Reviews.csv')
data
```

```python
dataset=data.drop(['Title','Review Text','Clothing ID','Unnamed: 0'],axis= 1)
print(dataset)
```

```python
dataset.isnull().sum()
```

```python
dataset = dataset.fillna('')
```

```python
print(dataset)
```

```python
dataset.isnull().sum()
```

```python
age_bins = [18,25,35,45,np.inf]
age_bins_label =['Young','Younger Adults','Adults','Senior Citizen']
```

```python
dataset_copy = dataset.copy()
dataset_copy['Category-Age'] = pd.cut(dataset['Age'],bins=age_bins,labels=age_bins_label)
```

```python
dataset_copy
```

```python
dataset
```

```python
customer_count = dataset_copy['Category-Age'].value_counts().sort_index()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(age_bins_label, customer_count)
plt.xlabel('Age Groups')
plt.ylabel('Customers')
plt.title('Customers V/S Age Groups')
plt.show()
```

```python
selected_columns = ['Age', 'Rating']
selected_data = dataset_copy[selected_columns]
covariance_matrix = selected_data.cov()
```

```python
plt.figure(figsize=(6, 6))
plt.imshow(covariance_matrix, cmap='coolwarm', interpolation='none')
plt.title("Covariance Matrix Heatmap")
plt.colorbar(label="Covariance")
plt.xticks(range(len(selected_columns)), selected_columns)
plt.yticks(range(len(selected_columns)), selected_columns)
plt.show()
```

```python
dataset_copy['Repeated Purchase'] = dataset_copy['Rating'] > 3
dataset_copy['Repeated Purchase'] = dataset_copy['Repeated Purchase'].apply(lambda x: 1 if x else 0)
```

```python
dataset_copy
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dataset_copy['Category-Age'] = le.fit_transform(dataset_copy['Category-Age'])
```

```python
dataset_copy
```

```python
from scipy.stats import chi2_contingency

chi_data= pd.crosstab(dataset_copy['Repeated Purchase'],dataset_copy['Category-Age'])


chi_stat, p_value, dof, expected = chi2_contingency(chi_data)

print("Chi-squared statistic:", chi_stat)
print("P-value:", p_value)
print("Degrees of freedom:", dof)
print("Expected frequencies table:")
print(expected)
```

```python
chi_data.plot(kind='bar', stacked=True)
plt.title("Repeated Purchase Vs Age Group")
plt.xlabel("Repeated Purchase")
plt.ylabel("Count")
plt.legend(title="Age Group")
plt.show()
```

```python
from scipy.stats import ttest_ind
repeated_purchase = dataset_copy[dataset_copy['Repeated Purchase'] == 1]['Rating']
not_repeated_purchase = dataset_copy[dataset_copy['Repeated Purchase'] == 0]['Rating']
t_statistic, p_value = ttest_ind(repeated_purchase, not_repeated_purchase)

print("T-statistic:", t_statistic)
print("P-value:", p_value)
```

```python
plt.figure(figsize=(10, 6))

plt.hist(repeated_purchase, bins=np.linspace(0, 5, 21), alpha=0.5, label='Repeated Purchase')
plt.hist(not_repeated_purchase, bins=np.linspace(0, 5, 21), alpha=0.5, label='Not Repeated Purchase')

plt.axvline(np.mean(repeated_purchase), color='blue', linestyle='dashed', linewidth=2, label='Mean Repeated Purchase')
plt.axvline(np.mean(not_repeated_purchase), color='orange', linestyle='dashed', linewidth=2, label='Mean Not Repeated Purchase')

plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Rating Distribution by Repeated Purchase')
plt.legend()
plt.grid(True)
plt.show()
```
